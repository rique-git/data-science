## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read CSV

```python
data = pd.read_csv("data_estagios.csv")
data.head()
```

## Data Analysis

```python
sns.histplot(data=data, x="Hospital", hue = "Hospital")
plt.title("Número de pacientes Estágio 1 por Local")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
sns.histplot(data=data, x="Dia", hue = "Hospital", discrete=True)
plt.title("Utentes por dia Estágio 1 por Local")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
sns.histplot(data=data, x="Sexo", hue = "Sexo")
plt.title("Número de utentes por sexo")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=data, x="Razão", hue = "Razão")
plt.title("Número de utentes por motivo da visita")
plt.xlabel("Motivo")
plt.ylabel("Número de Utentes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=data, x="Solução", hue = "Solução")
plt.title("Número de utentes por solução")
plt.xlabel("Solução")
plt.ylabel("Número de Utentes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=data, x="Famílias", hue = "Famílias")
plt.title("Número de utentes por Famílias")
plt.xlabel("Famílias")
plt.ylabel("Número de Utentes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data = data, x="Famílias", hue="Sexo")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data = data[data["Sexo"] == "Homem"], x="Famílias", palette="Blues", hue = "Famílias")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data = data[data["Sexo"] == "Mulher"], x="Famílias", palette="Oranges", hue = "Famílias")
plt.xticks(rotation=45)
plt.show()
```
