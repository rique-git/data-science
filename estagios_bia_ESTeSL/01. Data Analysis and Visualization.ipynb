## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read CSV

```python
data = pd.read_csv("data_estagios.csv")
data.head()
```

## Data Analysis

```python
sns.histplot(data=data, x="Local", hue = "Local")
plt.title("Número de Pacientes por Local de Estágio")
plt.ylabel("Número de Pacientes")
plt.legend().remove()
plt.show()
```

```python
data_1 = data[data["Estágio"] == 1].copy()
data_2 = data[data["Estágio"] == 2].copy()
data_3 = data[data["Estágio"] == 3].copy()
```

## Estágio 1 - Santa Maria + Pulido Valente (Super Ajudas)

```python
sns.histplot(data=data_1, x="Dia", hue = "Local", discrete=True)
plt.title("Utentes por dia Estágio 1 por Local")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
sns.histplot(data=data_1, x="Sexo", hue = "Sexo")
plt.title("Número de Pacientes por Sexo (estágio Super Ajudas)")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=data, x="Razão", hue = "Razão")
plt.title("Número de utentes por motivo da visita")
plt.xlabel("Motivo")
plt.ylabel("Número de Utentes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=data, x="Solução", hue = "Solução")
plt.title("Número de utentes por solução")
plt.xlabel("Solução")
plt.ylabel("Número de Utentes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data=data, x="Famílias", hue = "Famílias", palette="bright")
plt.title("Vendas (estágio Super Ajudas)", size = 20)
plt.xlabel("Famílias", size = 12)
plt.ylabel("Número de Utentes", size = 12)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data = data, x="Famílias", hue="Sexo")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data = data[data["Sexo"] == "Homem"], x="Famílias", palette="Blues", hue = "Famílias")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(data = data[data["Sexo"] == "Mulher"], x="Famílias", palette="Oranges", hue = "Famílias")
plt.xticks(rotation=45)
plt.show()
```

## Estágio 2 - APDP

```python
sns.histplot(data=data_2, x="Dia", hue = "Dia", discrete=True, palette="Blues")
plt.title("Utentes por dia Estágio 1 por Local")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
data_2["Sexo"] = pd.Categorical(data_2["Sexo"], categories=["Homem", "Mulher"], ordered=True)
sns.histplot(data=data_2, x="Sexo", hue = "Sexo")
plt.title("Número de utentes por sexo (estágio APDP)")
plt.ylabel("Número de Utentes")
plt.show()
```

## Estágio 3 - SIORTO

```python
data_3["Sexo"] = pd.Categorical(data_3["Sexo"], categories=["Homem", "Mulher"], ordered=True)
sns.histplot(data=data_3, x="Sexo", hue = "Sexo")
plt.title("Número de utentes por sexo (estágio SIORTO)")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
sns.histplot(data=data_3, x="Amputação", hue = "Amputação")
plt.title("Número de Utentes por Nível de Amputação (estágio SIORTO)")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
sns.histplot(data=data_3, x="Protetização", hue = "Protetização", palette= "hls")
plt.title("Número de Utentes por Protetização (estágio SIORTO)")
plt.ylabel("Número de Utentes")
plt.show()
```

```python
sns.histplot(data=data_3, x="Causa da amputação", hue = "Causa da amputação", palette= "viridis")
plt.title("Número de Utentes por Causa da Amputação (estágio SIORTO)")
plt.ylabel("Número de Utentes")
plt.show()
```
