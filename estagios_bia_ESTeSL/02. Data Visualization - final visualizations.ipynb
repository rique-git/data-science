## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read CSV

```python
data = pd.read_csv("data_estagios.csv")
data.head()
```

## Data Analysis

```python
ax = sns.histplot(data=data, x="Local", hue = "Local", legend=False)

for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')
            
y_max = max([bar.get_height() for container in ax.containers for bar in container])
ax.set_ylim(0, y_max * 1.15)

plt.title("Número de Pacientes por Local de Estágio")
plt.ylabel("Número de Pacientes")

plt.savefig("images/número_pacientes_por_local.png")
plt.show()
```

```python
data_1 = data[data["Estágio"] == 1].copy()
data_2 = data[data["Estágio"] == 2].copy()
data_3 = data[data["Estágio"] == 3].copy()
```
